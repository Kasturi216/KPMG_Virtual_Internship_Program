<h3 style='color:green;'><b> Customer Demographic Table </b></h3>

```python
path="https://cdn-assets.theforage.com/vinternship_modules/kpmg_data_analytics/KPMG_VI_New_raw_data_update_final.xlsx"
customer_data=pd.read_excel(path,sheet_name='CustomerDemographic',header=0)
customer_data.columns=customer_data.iloc[0,]
customer_data.drop(0,axis=0,inplace=True)
customer_data.head()
```

<b> We see that there is a column named 'default', which has inconceivable data. <br>
    Hence, we would like to refer back to the client for more knowledge. </b>

<b> Since, at this point of time the 'default' column makes no sense, so dropping it will not hamper out data quality. </b>

```python
customer_data.drop('default',axis=1,inplace=True)
```

```python
customer_data.info()
```

<b> We see that 'past_3_years_bike_related_purchases', 'DOB', 'tenure' attributes are in wrong data format. So, we need to change them into their respective correct format, which are 'int64', 'datetime', 'int64' respectively. </b>

```python
customer_data['past_3_years_bike_related_purchases']=customer_data['past_3_years_bike_related_purchases'].astype('int64')
customer_data['tenure']=customer_data['tenure'].astype('float')
customer_data['DOB']=customer_data['DOB'].astype('datetime64[ns]')
```

```python
from datetime import datetime, date
customer_data["age"]=round((datetime.now()-customer_data["DOB"]).dt.days/365,1)
customer_data.describe(include='all',datetime_is_numeric=True)
```

```python
wrong_age = customer_data[customer_data["age"]>104]
customer_data.drop(wrong_age.index[0],axis=0,inplace=True)
customer_data.reset_index(inplace=True,drop= True)
```

```python
customer_data.describe(include='all', datetime_is_numeric=True)
```

```python
customer_data['gender'].unique()
```

```python
customer_data['gender'] = customer_data['gender'].replace(['F','Femal'],'Female')
customer_data['gender'] = customer_data['gender'].replace(['M'],'Male')   
customer_data['gender'] = customer_data['gender'].replace(['U'],'Unspecified')
```

```python
print(customer_data['gender'].value_counts())
print('\n')
print(customer_data['gender'].unique())
```

```python
customer_data.isnull().sum()
```

<h3 style='color:green;'><b> Customer Address Table </b></h3>

```python
path="https://cdn-assets.theforage.com/vinternship_modules/kpmg_data_analytics/KPMG_VI_New_raw_data_update_final.xlsx"
customer_address=pd.read_excel(path,sheet_name='CustomerAddress',header=1)
customer_address
```

```python
customer_address.info()
```

<b> No null values are there in the data.<br>
    Also, The Datatype of the columns 'customer_id' and 'postcode' should rather be in object, so we convert them.</b>

```python
customer_address['customer_id']=customer_address['customer_id'].astype('str')
customer_address['postcode']=customer_address['postcode'].astype('str')
```

```python
customer_address.duplicated().sum()
```

<b> No duplicate values are there. </b>

```python
for colmn in customer_address.columns:
    if len(customer_address[colmn].unique())<10:
        print(f'Unique values of column {colmn}: ',customer_address[colmn].unique())
```

<b> We see that in the column 'state' there there are 'New South Wales', 'Victoria' are present as well as their abbreviations 'NSW' and 'VIC' respectively. So we would like to replace those values with their abbreviations so that we do not repeat them. </b>

```python
customer_address['state']=customer_address['state'].replace('New South Wales', 'NSW')
customer_address['state']=customer_address['state'].replace('Victoria', 'VIC')
print(customer_address['state'].value_counts())
print('\n')
print(customer_address['state'].unique())
```

```python
new_file = pd.ExcelWriter('CorrectedData.xlsx')

transaction_data.to_excel(new_file, sheet_name = 'transaction_data', index = False)
customer_data.to_excel(new_file, sheet_name = 'customer_data', index = False)
customer_address.to_excel(new_file, sheet_name = 'customer_address', index = False)

new_file.save()
```
